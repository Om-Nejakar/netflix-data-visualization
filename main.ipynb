```python
print(df.isnull().sum()) #getting all colns with sum of null values 
print('****************')
print(df.isnull().sum()[df.isnull().sum() > 0]) # apply cond to show that which are having nulls 
```

```python
#filled the NULL values with the unknown value
df['country'] = df['country'].fillna('unknown')
df['cast'] = df['cast'].fillna('not specified')
df['director'] = df['director'].fillna('not specified')
df['date_added'] = df['date_added'].fillna('unknown')

df.head(10)
```

```python
#verifing if any null cols remains 

print(df.isnull().sum()) #getting all colns with sum of null values 
print('****************')
print(df.isnull().sum()[df.isnull().sum() > 0]) # apply cond to show that which are having nulls 
```

```python
#converting the string date and time into datetime data type so that we can use it more efficient way 

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce') # for error handling errors='coerce'

#so that usefull function can be added 
print(df['date_added'].dt.year)
print(df['date_added'].dt.month)
df.head()
```

```python
# standardizing the text 

df['listed_in'] = df['listed_in'].str.lower()
df['title'] = df['title'].str.lower()
df.head()
```

```python
# checking the m*n matrix
df.shape
```

```python
import matplotlib.pyplot as plt

# Count Movies and TV Shows
movie_sum = (df['type'] == 'Movie').sum()
tvshow_sum = (df['type'] == 'TV Show').sum()

# X and Y values
X = ['Movies', 'TV Shows']
Y = [movie_sum, tvshow_sum]
bar_colors = ['green', 'yellow']

# Create bar chart
bars = plt.bar(
    X, Y,
    width=0.5,
    edgecolor='black',
    linewidth=2
)

# Set colors for each bar
for bar, color in zip(bars, bar_colors):
    bar.set_color(color)

# Add numbers on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # X position (center of bar)
        yval + 100,                       # Y position (just above the bar)
        int(yval),                        # The actual value
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

# Labels and title
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.title("Movies vs TV Shows on Netflix")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df.head(1)
df['year_added'] = df['date_added'].dt.year
counts_per_year = df['year_added'].value_counts().sort_index() #this is basically a series 

plt.plot(counts_per_year.index, counts_per_year.values, marker='o', linewidth=2)
plt.show()
```

```python
# most country contributated to netflix

df['country'] = df['country'].str.split(',') #making into array of each country 
df = df.explode('country') # making each ele in array to rows
df['country'] = df['country'].str.strip()  # removing any extra space if there 
country_content = df["country"].value_counts() # count the countries 
top10_countries = df['country'].value_counts().head(10).sort_values(ascending=True) # top 10 in ascending so that low value get y lowest values first 
# top10_countries = top10_countries.iloc[::-1] # reversing 
plt.barh(top10_countries.index, top10_countries.values)
plt.show()
```

```python
# most common genre 
df["listed_in"] = df["listed_in"].str.split(',')
df = df.explode("listed_in")
df["listed_in"] = df["listed_in"].str.strip()
top10_genre = df["listed_in"].value_counts().head(10).sort_values(ascending=True)

plt.barh(top10_genre.index, top10_genre.values)
plt.show()
```

```python
df["duration"]
movies = df[df['type'] == 'Movie'].copy()
shows = df[df['type'] == 'TV Show'].copy()


movies["duration_min"] = movies["duration"].str.replace(" min", "").astype(float)
shows["seasons"] = shows["duration"].str.replace(" Season", "").str.replace("s", "").astype(int)
shows["seasons_num"] = shows["duration"].str.extract("(\d+)").astype(int)

print(shows["seasons"].head(10).index)
print(shows["seasons"].head(10).values)

fig , axes = plt.subplots(1,2, figsize=(12,6))

axes[0].bar(movies["duration_min"].head(10).index, movies["duration_min"].head(10).values)
axes[1].bar(shows["seasons_num"].head(10).index, shows["seasons_num"].head(10).values)
```

```python
directors = df[df["director"] != "not specified"]
top10_directors = directors["director"].value_counts().sort_values(ascending=False).head(10)
print(top10_directors)

#seaborn basically making it more good visual  
sns.barplot(x=top10_directors.values, y=top10_directors.index, palette="viridis")
# plt.barh(top10_directors.index, top10_directors.values)   
plt.show()
```
